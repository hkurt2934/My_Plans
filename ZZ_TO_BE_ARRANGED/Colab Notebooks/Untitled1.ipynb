```
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import random
from pandas import DataFrame
Data_Frame = {'x': [random.randint(0,100) for i in range(100)],
             'y': [random.randint(0,100) for i in range(100)]}
df = DataFrame(Data_Frame, columns = ['x','y'])
kmeans = KMeans (n_clusters = 3)
kmeans.fit(df)
merkezler  = kmeans.cluster_centers_
print(merkezler)
plt.scatter(df['x'],df['y'], c = kmeans.labels_.astype(float))
plt.scatter(merkezler[:,0], merkezler[:,1], c = 'red')
```
