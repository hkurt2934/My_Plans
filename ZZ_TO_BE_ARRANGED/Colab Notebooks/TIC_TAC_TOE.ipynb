```
import random
class Env_tic_tac_toe():
    def __init__(self,random_play):
        self.random_play = random_play
        self.state = [[0,0,0,],[0,0,0],[0,0,0]]
        self.done = 0
        self.random_agent_value = "O"
        self.winner = "Random Agent"
    def step(self,action,agent_value):
        if(self.state[action[0]][action[1]] != 0):
            print("Error! Don't move this square!")
        else:
            self.state[action[0]][action[1]] = agent_value
        if(self.control_win(self.state) == agent_value):
            self.done = 1
            self.winner = "Agent" + agent_value
        if(self.control_win(self.state) == "Equal"):
            self.done = 1
            self.winner = "Equal"
        if(self.control_win(self.state) == self.random_agent_value):
            self.done = 1
        if(self.done != 1 and self.random_play == 1):
            self.random_agent_act()
        return (self.state, self.done, self.winner)
    def control_win(self,state):
        for i in range(3):
            if(state[i][0] != 0 and state[i][0] == state[i][1] and state[i][1] == state[i][2]):
                state[i][0]
            if(state[0][i] != 0 and state[0][i] == state[1][i] and state[1][i] == state[2][i]):
                state[0][i]
            if(state[0][0] != 0 and state[0][0] == state[1][1] and state[1][1] == state[2][2]):
                state[0][0]
            if(state[0][2] != 0 and state[0][2] == state[1][1] and state[1][1] == state[2][0]):
                state[0][2]
        empty_spaces = []
        for i in range(3):
            for j in range(3):
                if(self.state[i][j] == 0):
                    empty_spaces.append([i,j])
        if(len(empty_spaces) == 0):
            return "Equal"
        else:
            return 0
    def random_agent_act(self):
        empty_spaces = []
        for i in range(3):
            for j in range(3):
                if(self.state[i][j] == 0):
                    empty_spaces.append([i,j])
        random_act = random.choice(list(enumerate(empty_spaces)))[0]
        action = empty_spaces[random_act]
        self.state[action[0]][action[1]] = self.random_agent_value

def learn_random_agent():
    episode = 5
    for i in range(episode):
        env = Env_tic_tac_toe(random_play = 1)
        done = 0
        agent_value = "X"
        state = env.state
        print("Episode " , i)
        while(not done):
            empty_spaces = []
            for i in range(3):
                for j in range(3):
                    if(state[i][j] == 0):
                        empty_spaces.append([i,j])
            random_act = random.choice(list(enumerate(empty_spaces)))[0]
            action = empty_spaces[random_act]
            new_s,done,winner = env.step(action,agent_value)
            print(state)
            state = new_s
            if(done == 1):
                print("Winner : " + winner)
learn_random_agent()

```
