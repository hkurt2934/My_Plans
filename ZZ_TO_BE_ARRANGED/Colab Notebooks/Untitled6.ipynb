```
import random as rnd 
random_number = rnd.randint(1,10)
count = 0

while(True):
  enter_number = input("Enter a Number :")
  count+=1
  if(enter_number == "Exit"):
    break
  enter_number = int(enter_number)
  if(enter_number > random_number):
    print("Please, Enter smaller number :")
  elif(enter_number < random_number):
    print("Please, Enter higher number :")
  else:
    print("You win:)")
    print("Your Score : ", count)
    break
```
